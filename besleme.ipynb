```python
import os
import shutil
import time
import random

kaynak_ok = r"C:\Users\safak\OneDrive\Masaüstü\Personal Projects\ML_visual\Veri_Seti\test\ok_front"
kaynak_nok = r"C:\Users\safak\OneDrive\Masaüstü\Personal Projects\ML_visual\Veri_Seti\test\def_front"

# Arayüzün izlediği (kamera) hedef klasörler
hedef_ch1 = r"C:\Users\safak\KAMERA_1_GELEN"
hedef_ch2 = r"C:\Users\safak\KAMERA_2_GELEN"

# Test edilecek toplam parça sayısı
TEST_ADEDI = 100 
BEKLEME_SURESI = 0.5 # İki parça arası bekleme süresi(500 ms)

# DOSYA HAZIRLAMA
ok_dosyalar = [os.path.join(kaynak_ok, f) for f in os.listdir(kaynak_ok) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
nok_dosyalar = [os.path.join(kaynak_nok, f) for f in os.listdir(kaynak_nok) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
tum_dosyalar = ok_dosyalar + nok_dosyalar
random.shuffle(tum_dosyalar)

# İstenen sayıda parçayı al
test_havuzu = tum_dosyalar[:TEST_ADEDI]

test_havuzu = tum_dosyalar[:TEST_ADEDI]

print(" ÇİFT KANALLI SANAL BESLEYİCİ ÇALIŞTIRILDI ")
print("-" * 100)
print(f"Toplam {len(test_havuzu)} adet parça (OK/NOK karışık), rastgele CH1 ve CH2'ye veriliyor...\n")

# 50/50 DAĞILIM 
yari_sayi = TEST_ADEDI // 2
hedef_torbasi = [hedef_ch1] * yari_sayi + [hedef_ch2] * (TEST_ADEDI - yari_sayi)
random.shuffle(hedef_torbasi) # CH1 ve CH2 sırasını rastgele karıştır

# ÜRETİM BANDI SİMÜLASYONU 
for i, kaynak_yol in enumerate(test_havuzu, 1):
    dosya_adi = os.path.basename(kaynak_yol)
    
    hedef_klasor = hedef_torbasi[i-1]
    kanal_adi = "CH1" if hedef_klasor == hedef_ch1 else "CH2"
    yeni_dosya_adi = f"simtest_{i}_{dosya_adi}"
    hedef_yol_yeni = os.path.join(hedef_klasor, yeni_dosya_adi)
    
    shutil.copy(kaynak_yol, hedef_yol_yeni)    
    durum = "OK " if kaynak_yol in ok_dosyalar else "NOK"
    print(f"[{i}/{len(test_havuzu)}] -> {kanal_adi} Bandı | Parça: {durum} | {yeni_dosya_adi}")
    time.sleep(BEKLEME_SURESI) 

print("-" * 100)
print("Test beslemesi tamamlandı! Lütfen uygulamanın log sekmesinden E2E sürelerini inceleyin.")
```
